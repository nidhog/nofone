```python
import utils.glucose as gl
import utils.prep as p
from utils.plotting import resize
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
glucose = pd.read_csv('data-fasting/glucose.csv')
glucose.nunique()
```

```python
dtime_lbl='dtime'
g_lbl = 'Historic Glucose mmol/L'
glucose['dtime']=pd.to_datetime(glucose['Device Timestamp'], format='%d-%m-%Y %H:%M')
glucose.set_index(dtime_lbl, inplace=True)
```

```python
g_period = glucose['2020-03-15':'2020-05-30']
```

```python
resize()
plt.plot(g_period['2020-04-24':'2020-05-24'][g_lbl], color='red')
plt.plot(g_period['2020-05-25':][g_lbl], color='blue')

plt.xticks(rotation=45)
plt.show()
```

```python
resize()

plt.plot(g_period['2020-05-23'].index.map(lambda x:x.hour+(x.minute)/60), g_period['2020-05-23'][g_lbl], color='red', label='fasting')
#plt.plot(g_period['2020-05-23':'2020-05-24'].index[:127], g_period['2020-05-24'][g_lbl], color='blue', label='refeed')
plt.plot(g_period['2020-05-24'].index.map(lambda x:x.hour+(x.minute)/60), g_period['2020-05-24'][g_lbl], color='blue', label='fasting')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
